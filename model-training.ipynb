```python
# see results of tensorflow model
from tensorflow.keras.layers import Input, Dense, Activation, Dropout
from tensorflow.keras.models import Model
```

```python
# yoga_df.shape[1]
```

```python
input_layer = Input(shape=(yoga_df.shape[1],))
```

```python
dense_layer_1 = Dense(15, activation='relu')(input_layer)
```

```python
dense_layer_2 = Dense(10, activation='relu')(dense_layer_1)
```

```python
output = Dense(5, activation='softmax')(dense_layer_2)
```

```python
model = Model(inputs=input_layer, outputs=output)
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
print(model.summary())
```

```python
history = model.fit(features_train,labels_train, batch_size=8,epochs=50,verbose=1,validation_split=0.35)
```

```python
score = model.evaluate(features_test, labels_test, verbose=1)
```

```python
print("test score: ",  score[0])
```
